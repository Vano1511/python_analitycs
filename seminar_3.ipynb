```python
import pandas as pd
```

```python
df = pd.read_csv("./laptop_price.csv", encoding="latin-1")
df
```

```python
df["Cpu_company"] = df["Cpu"].apply(lambda x: x.split()[0])
df["Cpu_company"].head(15)
```

```python
df.head()
```

```python
def repack_memory_in_gb(string: str) -> int:
    list_string = string.split()
    return_list = []
    for el in list_string:
        if el.endswith("TB"):
            return_list.append(int(float(el[:-2]) * 1024))
        elif el.endswith("GB"):
            return_list.append(int(float(el[:-2])))
    return sum(return_list)
    
df["Memory_amount"] = df["Memory"].apply(repack_memory_in_gb)
df.sample(10)
```

```python
df = df.drop(columns=["Memory_amount", "ScreenResolution"])
df.sample(10)
```

```python
df["SSD"] = 0
df.sample(10)
```
