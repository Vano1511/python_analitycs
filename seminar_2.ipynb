```python
import pandas as pd
```

```python
df = pd.read_csv("./laptop_price.csv", encoding="latin-1")
df.describe(include="object")
```

```python
df.sort_values(["Price_euros"], ascending=False)
```

```python
df[df.Inches == df.Inches.min()][["Company", "Inches", "TypeName", "Price_euros"]]
```

```python
df[(df["Company"] == "HP")].sort_values(["Price_euros"], ascending=False).head()
```

```python
ultrabooks_8RAM_count = df[(df["TypeName"] == "Ultrabook") & (df["Ram"] == "8GB")].count()[0]
ultrabooks_8RAM_count
```

```python
all_ultrabooks_count = df[(df["TypeName"] == "Ultrabook")].count()[0]
all_ultrabooks_count
```

```python
part_8gbram_of_ultrabooks = f"{round(ultrabooks_8RAM_count / all_ultrabooks_count * 100, 2)}%"
part_8gbram_of_ultrabooks
```

```python
df[(df["OpSys"] == "Windows 10") & (df["Ram"].isin(("8GB", "16GB"))) & (df["Price_euros"] < 500)].shape[0]
```

```python
minim = df[
           (df["Company"] == "MSI") &
           (df["Gpu"] == "Nvidia GeForce GTX 1050 Ti") &
           (~(df["Inches"] == 15.6))
].Price_euros.min()
maxim = df[
    (df["Company"] == "MSI") & 
    (df["Gpu"] == "Nvidia GeForce GTX 1050 Ti") &
    (~(df["Inches"] == 15.6))
].Price_euros.max()
(minim, maxim)
```

```python
avg_7600 = df[df["Cpu"] == "Intel Core i7 7600U 2.8GHz"].Price_euros.mean()
avg_7700 = df[df["Cpu"] == "Intel Core i7 7700HQ 2.8GHz"].Price_euros.mean()
avg_7600, avg_7700
```
